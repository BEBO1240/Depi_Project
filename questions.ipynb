```python
import pandas as pd
file_path = "Data.xlsx"
df = pd.read_excel(file_path)
df.head()
```

```python
top_cities = df.groupby(['Region', 'City'])['Customer ID'].nunique().reset_index(name='unique_customers')
top_cities = top_cities.sort_values(['Region', 'unique_customers'], ascending=[True, False])
top_3_cities = top_cities.groupby('Region').head(3)
top_3_cities
```

```python
order_freq = df.groupby(['Segment', 'Customer ID']).size().reset_index(name='order_count')
avg_order_freq = order_freq.groupby('Segment')['order_count'].mean().sort_values(ascending=False)
print("\nQ2 - Customer segments by average order frequency:\n", avg_order_freq)
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print("\nQ3 - Top 10 products by sales:\n", top_products)
```

```python
product_count = df.groupby('Sub-Category')['Product Name'].nunique()
print("\nQ4 - Number of products in each sub-category:\n", product_count)
```

```python
df['year_month'] = df['Order Date'].dt.to_period('M')
monthly_orders = df.groupby(['City', 'year_month']).size().reset_index(name='orders')
monthly_orders['orders_lag'] = monthly_orders.groupby('City')['orders'].shift(1)
monthly_orders['growth_rate'] = (monthly_orders['orders'] - monthly_orders['orders_lag']) / monthly_orders['orders_lag']
fastest_growing = monthly_orders.groupby('City')['growth_rate'].mean().sort_values(ascending=False)
print("\nQ5 - Cities with fastest-growing order volume:\n", fastest_growing.head(10))
```

```python
df['quarter'] = df['Order Date'].dt.to_period('Q')
orders_by_quarter = df.groupby('quarter').size().sort_values(ascending=False)
print("\nQ6 - Quarters with highest order volume:\n", orders_by_quarter)
```

```python
orders_by_shipmode = df.groupby('Ship Mode').size()
print("\nQ7 - Orders by Ship Mode:\n", orders_by_shipmode)
```

```python
df['month'] = df['Order Date'].dt.to_period('M')
sales_by_month = df.groupby('month')['Sales'].sum().sort_values(ascending=False)
print("\nQ8 - Month with highest total sales:\n", sales_by_month.head(1))
```

```python
order_sales = df.groupby('Order ID')['Sales'].sum().sort_values(ascending=False).head(10)
print("\nQ9 - Top 10 orders with highest sales:\n", order_sales)
```
