```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Superstore Sales Dataset.csv")
df
```

```python
df = df[['Customer ID','Order ID', 'Order Date', 'Ship Date', 'Ship Mode','Region','State','City']]
```

```python
df.isnull().sum()
```

```python
df = df.copy()
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y', errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y', errors='coerce')
```

```python
df.to_csv('Order_Data.csv', index=False)
```

```python

orders_per_customer = df.groupby('Customer ID')['Order ID'].count().sort_values(ascending=False)
orders_per_customer.head(25)
```

```python

orders = df.groupby([df['Order Date'].dt.year.sort_values(ascending = False), df['Order Date'].dt.month_name().sort_values(ascending = False)])['Order ID'].count()
orders

```

```python
ship_mode_counts = df.groupby('Ship Mode')['Order ID'].count().sort_values(ascending = False)
ship_mode_counts
```
